## Set up

```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline
```

```python
df = pd.read_csv('traffic-collision-data-from-2010-to-present.csv')
df
```

```python
df.drop(['Area ID','Crime Code','DR Number',
                   'MO Codes','Premise Code','Precinct Boundaries','LA Specific Plans'],axis=1)
df['Date'] =  pd.to_datetime(df['Date Occurred'], format='%Y-%m-%dT%H:%M:%S')
df['year'] = pd.DatetimeIndex(df['Date']).year
df['month'] = pd.DatetimeIndex(df['Date']).month
df['weekday']= pd.DatetimeIndex(df['Date']).dayofweek
```

## Which area reported the most collisions? 

```python
df = df[df['Address'].notnull()]
df =df[df['Cross Street'].notnull()]
```

```python
df['Area Name'].value_counts().plot(kind='bar')
```

Let's test to see if 'Area ID' is identical to the data in 'Area Name'

```python
df['Area Name'].value_counts().plot(kind='bar')
```

```python
df['Area Name'].value_counts().values == df['Area ID'].value_counts().values
```

We can conclude these are identical.

Create a new column that has intersections of collisions using 'Address' and 'Cross Street'

```python
df['Address Cross Street']=((df['Address']+' '+df['Cross Street'])).str.replace(' +',' ')
```

```python
df['Address Cross Street'].value_counts()[0:30].plot(kind='bar')
```

Looking at the plot above, there is a problem: "Slauson and Western" should be the same intersection as "Western and Slauson", but these values show up differently. 

```python
df['Address'] = df['Address'].str.replace(' +',' ')
df['Cross Street'] = df['Cross Street'].str.replace(' +',' ')
df['Address Cross Street'] = df['Address'] +'*'+df['Cross Street']

def street_order(addresses):
    return " ".join(sorted(addresses.split('*')))

df['Address Cross Street'] = df['Address Cross Street'].map(street_order)
df['Address Cross Street'][0:10]
```

```python
df['Address Cross Street'].value_counts()[0:30].plot(kind='bar')
```

### Within each Council District, which Neighborhood Councils have the most collisions? 

```python
df['Unique Collision'] = df['DR Number'].astype(str) + df['Address Cross Street']
by_nbhd = df.groupby(by = ['Council Districts','Neighborhood Councils (Certified)'])[['Unique Collision']].agg(['count']).unstack(['Neighborhood Councils (Certified)'])
max_col = by_nbhd.idxmax(axis=1)
max_col
```

This means that for council district 1, the neighborhood council 11 had the most collisions. 
For council district 2, neighborhood council 43 had the most collisions, etc...
